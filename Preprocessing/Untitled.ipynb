```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('houses.csv')
```

```python
df.head()
```

```python
print(df.isnull().sum())
df=df.dropna()
```

```python
df.isnull().sum()
print(df.shape)
```

```python
from ydata_profiling import ProfileReport
pr=ProfileReport(df)
pr.to_file(output_file='analysis.html')
```

```python
x=df.iloc[:,0:17]
y=df.iloc[:,17:18]

y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x_train,y_train)
y_pre = model1.predict(x_test)
r2_score(y_test,y_pre)
```

```python
score=cross_val_score(model1,y_test,y_pre,cv=10)
print(score.mean())
```

```python
trnf1 = ColumnTransformer([
    ('first',FunctionTransformer(np.log),[1,2,3,4,9,11,15,16])
],remainder='passthrough')
```

```python
x_trans=trnf1.fit_transform(x_train)
```

```python
pd.DataFrame(x_trans).isnull().sum()
```

```python
model2=LinearRegression()
model2.fit(x_trans,y_train)
y_pre = model2.predict(x_test)
r2_score(y_test,y_pre)
```
